# 3.Process

### 3.1 Missing Values

```python
# Number of missing data

missing_values= data.isna().sum()
missing_values
```

**Note 4: For resolving this problem we have differents alternatives in my case i drop this values in other notebook i used Clean Data Analysis**

```python
# Percentage of data missing

total_cells = np.product(data.shape)
total_missing_values = missing_values.sum()

missing_percent = (total_missing_values/total_cells)*100
print(f'{total_missing_values} out of {total_cells} ({missing_percent:.2f}%) missing')
```

```python
# Drop NA Values

data = data.dropna(how='any')\
    .reset_index(drop=True)\
    .copy()
```

```python
data.info()
```

### 3.2 Duplicated values

```python
#This colums has duplicated values

data[data.duplicated()]
```

```python
data.loc[data.duplicated(subset=['Name'])]
```

**Note:We check the name of videogames is the same but Platform is different**

```python
data.loc[data.duplicated(subset=['Name','Platform'])]
```

**Note: After research, "Need for Speed: Most Wanted" have two versions in 2005 and 2012, therefore, I will rename these values.**

```python
data.loc[1016,'Name'] = 'Need for Speed: Most Wanted 2005'
data.loc[5675,'Name'] = 'Need for Speed: Most Wanted 2012'
data.loc[[1016, 5675],'Name']
```

```python
data.loc[data['Name'] == 'Madden NFL 13']
```

**Note: Row 408 and row 6755 are same game and on same platform, confirm they're duplicated but index 6755 present errors**

```python
# Drop row 6755

data = data.loc[~data.duplicated(subset=['Name','Platform'])].reset_index(drop=True).copy()
```

```python
data.loc[data.duplicated(subset=['Name','Platform'])]
```

### 3.3 Inconsistent Data

```python
check1 = data.groupby(['Name','Platform','Year_of_Release','Genre','Publisher','Developer','Rating'])['Name'].count().to_frame()

check1.query('Name > 1')
```

```python
check2 = data.groupby(['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales',
                            'Other_Sales', 'Global_Sales', 'Critic_Score',
                            'Critic_Count', 'User_Score', 'User_Count'])['Name'].count().to_frame()
check2.query('Name > 1')
```
